2. Thống kê tổng số đơn hàng, số lượng khách hàng và người bán.


### Thống kê tổng quan về đơn hàng, khách hàng và người bán

```python
from pyspark.sql.functions import countDistinct

# Thống kê tổng số đơn hàng
total_orders = orders_df.select(countDistinct('Order_ID')).collect()[0][0]
print(f"Tổng số đơn hàng: {total_orders}")

# Thống kê tổng số khách hàng
total_customers = customer_list_df.select(countDistinct('Customer_Trx_ID')).collect()[0][0]
print(f"Tổng số khách hàng: {total_customers}")

# Thống kê tổng số người bán
total_sellers = order_items_df.select(countDistinct('Seller_ID')).collect()[0][0]
print(f"Tổng số người bán: {total_sellers}")
```

3. Phân tích số lượng đơn hàng theo quốc gia, sắp xếp theo thứ tự giảm dần.


### Phân tích số lượng đơn hàng theo quốc gia

```python
from pyspark.sql.functions import col

# Join orders_df with customer_list_df on Customer_Trx_ID
orders_with_country_df = orders_df.join(customer_list_df, on='Customer_Trx_ID', how='inner')

# Group by Customer_Country and count distinct orders
orders_by_country = orders_with_country_df.groupBy('Customer_Country').agg(
    countDistinct('Order_ID').alias('Total_Orders')
)

# Sort in descending order of Total_Orders
orders_by_country_sorted = orders_by_country.orderBy(col('Total_Orders').desc())

print('Số lượng đơn hàng theo quốc gia (sắp xếp giảm dần):')
orders_by_country_sorted.show(10)
```

4. Phân tích số lượng đơn hàng nhóm theo năm, tháng đặt hàng (Hiển thị theo năm
tăng dần, tháng giảm dần)

### Phân tích số lượng đơn hàng theo năm, tháng đặt hàng

```python
from pyspark.sql.functions import year, month, countDistinct, col

# Extract year and month from 'Order_Purchase_Timestamp'
orders_df_with_ym = (
    orders_df
    .withColumn('Order_Year', year(col('Order_Purchase_Timestamp')))
    .withColumn('Order_Month', month(col('Order_Purchase_Timestamp')))
)

# Group by year and month, and count distinct orders
orders_by_year_month = orders_df_with_ym.groupBy('Order_Year', 'Order_Month').agg(
    countDistinct('Order_ID').alias('Total_Orders')
)

# Sort by year ascending, then month descending
orders_by_year_month_sorted = orders_by_year_month.orderBy(col('Order_Year').asc(), col('Order_Month').desc())

print('Số lượng đơn hàng theo năm, tháng đặt hàng (năm tăng dần, tháng giảm dần):')
orders_by_year_month_sorted.show()
```

5. Thống kê điểm đánh giá trung bình và số lượng đánh giá theo từng mức

### Phân tích điểm đánh giá sản phẩm

```python
from pyspark.sql.functions import col, avg, count, when, expr
from pyspark.sql.types import IntegerType

# 1. Handle NULL values and convert Review_Score to integer
# Use expr("try_cast(Review_Score as INT)") to handle malformed strings gracefully
cleaned_reviews_df = order_reviews_df.withColumn(
    'Review_Score_Int',
    expr("try_cast(Review_Score as INT)")
)

# Filter out rows where conversion failed (Review_Score_Int is NULL)
# and ensure scores are between 1 and 5
cleaned_reviews_df = cleaned_reviews_df.filter(
    (col('Review_Score_Int').isNotNull()) &
    (col('Review_Score_Int').between(1, 5))
)

print('Schema sau khi làm sạch và chuyển đổi kiểu dữ liệu:')
cleaned_reviews_df.printSchema()
cleaned_reviews_df.show(5)

# 2. Calculate the average review score
average_review_score = cleaned_reviews_df.agg(avg('Review_Score_Int')).collect()[0][0]
print(f"\nĐiểm đánh giá trung bình: {average_review_score:.2f}")

# 3. Count reviews by score level (1 to 5)
review_counts_by_score = cleaned_reviews_df.groupBy('Review_Score_Int').agg(count('Review_Score_Int').alias('Total_Reviews'))

# Order by review score for better readability
review_counts_by_score_sorted = review_counts_by_score.orderBy(col('Review_Score_Int').asc())

print('\nSố lượng đánh giá theo từng mức điểm:')
review_counts_by_score_sorted.show()
```

6. Tính doanh thu (giá sản phẩm + phí vận chuyển) trong năm 2024 và nhóm theo danh mục sản phẩm

### Tính doanh thu theo danh mục sản phẩm trong năm 2024

```python
from pyspark.sql.functions import year, sum, col

# Join orders_df and order_items_df
orders_items_joined = orders_df.join(order_items_df, on='Order_ID', how='inner')

# Further join with products_df to get product category
all_data_joined = orders_items_joined.join(products_df, on='Product_ID', how='inner')

# Filter for orders in 2024
orders_2024_df = all_data_joined.filter(year(col('Order_Purchase_Timestamp')) == 2024)

# Calculate revenue for each item and sum by product category
revenue_by_category_2024 = orders_2024_df.withColumn(
    'Revenue', col('Price') + col('Freight_Value')
).groupBy('Product_Category_Name').agg(
    sum('Revenue').alias('Total_Revenue_2024')
).orderBy(col('Total_Revenue_2024').desc())

print('Doanh thu theo danh mục sản phẩm trong năm 2024 (sắp xếp giảm dần):')
revenue_by_category_2024.show(truncate=False)
```

7. Xác định sản phẩm có số lượng bán ra cao nhất và tính điểm đánh giá trung bình cho từng sản phẩm

### Sản phẩm bán chạy nhất

```python
from pyspark.sql.functions import count, desc

# Count the number of times each product_id appears in order_items_df
best_selling_products = order_items_df.groupBy('Product_ID').agg(
    count('Product_ID').alias('Total_Sold')
).orderBy(desc('Total_Sold'))

# Join with products_df to get product category name
best_selling_products_with_name = best_selling_products.join(products_df, on='Product_ID', how='left')

print('Top 10 sản phẩm bán chạy nhất:')
best_selling_products_with_name.show(10, truncate=False)
```

### Điểm đánh giá trung bình cho từng sản phẩm

```python
from pyspark.sql.functions import avg

# Join order_items_df with cleaned_reviews_df on Order_ID
products_with_reviews = order_items_df.join(cleaned_reviews_df, on='Order_ID', how='inner')

# Group by Product_ID and calculate the average review score
product_average_reviews = products_with_reviews.groupBy('Product_ID').agg(
    avg('Review_Score_Int').alias('Average_Review_Score')
).orderBy(desc('Average_Review_Score'))

# Join with products_df to get product category name
product_average_reviews_with_name = product_average_reviews.join(products_df, on='Product_ID', how='left')

print('Top 10 sản phẩm có điểm đánh giá trung bình cao nhất:')
product_average_reviews_with_name.show(10, truncate=False)
```

## 8. Tính toán hiệu suất giao hàng

```python
from pyspark.sql.functions import col, datediff

# Join orders_df with order_items_df on Order_ID
# Use 'all_data_joined' which already has orders and order_items joined, and also products. This is fine as we only need Order_Delivered_Carrier_Date and Shipping_Limit_Date
# Let's use 'orders_items_joined' for simplicity as we don't need product info for this task.
delivery_performance_df = orders_items_joined.withColumn(
    'Delivery_Lag_Days',
    datediff(col('Order_Delivered_Carrier_Date'), col('Shipping_Limit_Date'))
)

print('Hiệu suất giao hàng (Hiệu số giữa ngày giao hàng thực tế và ngày giao hàng dự kiến):')
# Show first 10 rows with selected columns for clarity
delivery_performance_df.select(
    'Order_ID',
    'Order_Delivered_Carrier_Date',
    'Shipping_Limit_Date',
    'Delivery_Lag_Days'
).show(10)
```

## 9. Nhóm khách hàng dựa trên số lượng đơn hàng, giá trị trung bình của đơn hàng và tần suất mua sắm

```python
from pyspark.sql.functions import col, sum, avg, countDistinct, min, max, datediff, when

# 1. Tính tổng giá trị cho mỗi đơn hàng (tổng tất cả các mặt hàng trong một đơn hàng)
#    orders_items_joined đã chứa Customer_Trx_ID và Order_Purchase_Timestamp
order_total_value_df = orders_items_joined.groupBy('Order_ID', 'Customer_Trx_ID', 'Order_Purchase_Timestamp').agg(
    sum(col('Price') + col('Freight_Value')).alias('Order_Total_Value')
)

# 2. Tổng hợp các chỉ số theo khách hàng
customer_segmentation_df = order_total_value_df.groupBy('Customer_Trx_ID').agg(
    countDistinct('Order_ID').alias('Total_Orders'),
    avg('Order_Total_Value').alias('Average_Order_Value'),
    min('Order_Purchase_Timestamp').alias('First_Purchase_Date'),
    max('Order_Purchase_Timestamp').alias('Last_Purchase_Date')
)

# 3. Tính tần suất mua hàng (số ngày giữa lần mua đầu tiên và cuối cùng)
customer_segmentation_df = customer_segmentation_df.withColumn(
    'Days_Active',
    datediff(col('Last_Purchase_Date'), col('First_Purchase_Date')) + 1
)

# Xử lý các trường hợp Days_Active là 1 (mua hàng một lần) để tránh chia cho số 0
# Nếu Days_Active là 1, tần suất mua hàng đơn giản là Total_Orders (là 1).
# Ngược lại, là Total_Orders / Days_Active.
# Điều này cho ra tỷ lệ đơn hàng trên mỗi ngày hoạt động.
customer_segmentation_df = customer_segmentation_df.withColumn(
    'Purchase_Frequency_Rate',
    when(col('Days_Active') == 1, col('Total_Orders'))
    .otherwise(col('Total_Orders') / col('Days_Active'))
)

# Chọn các cột liên quan để phân cụm/phân đoạn khách hàng
customer_segmentation_final_df = customer_segmentation_df.select(
    'Customer_Trx_ID',
    'Total_Orders',
    'Average_Order_Value',
    'Purchase_Frequency_Rate'
)

print('Chỉ số phân khúc khách hàng (Top 10):')
customer_segmentation_final_df.orderBy(col('Total_Orders').desc()).show(10)
```

## 10. Xếp hạng các seller dựa trên tổng doanh thu và số lượng đơn hàng bán được

```python
from pyspark.sql.functions import sum, countDistinct, col, desc

# Calculate Total Revenue and Total Orders Sold per Seller
seller_performance_df = orders_items_joined.groupBy('Seller_ID').agg(
    sum(col('Price') + col('Freight_Value')).alias('Total_Revenue'),
    countDistinct('Order_ID').alias('Total_Orders_Sold')
)

# Rank sellers by Total Revenue (descending) and then by Total Orders Sold (descending)
seller_ranked_df = seller_performance_df.orderBy(desc('Total_Revenue'), desc('Total_Orders_Sold'))

print('Top 10 Sellers by Revenue and Orders Sold:')
seller_ranked_df.show(10, truncate=False)
```
